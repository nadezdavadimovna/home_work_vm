### 2. Американские горки

```python
import numpy as np

def f1(x):
    return 2 * x**2 - np.log(x)

def f2(x):
    return x**3 / 3 + x**2
eps = 1e-6
a = 0.1  # Изменено, чтобы избежать log(0)
b = 1
while (b - a) > eps:
    c = (b + a) / 2
    f_left = f1(c - eps)
    f_right = f1(c + eps)
    if f_left < f_right:
        b = c
    else:
        a = c
print('Минимум f1 (дихотомия):', (a + b) / 2)

a = -1
b = 0.5
N = 30  
fib = [1, 1]
for i in range(2, N + 1):
    fib.append(fib[i - 1] + fib[i - 2])

x1 = a + (b - a) * fib[N - 2] / fib[N]
x2 = a + (b - a) * fib[N - 1] / fib[N]
f_1 = f2(x1)
f_2 = f2(x2)

for i in range(N - 2, 0, -1):
    if (x2 - x1) < 1e-8:  
        break
    if f_1 > f_2:
        a = x1
        x1 = x2
        x2 = b - (x1 - a)
        f_1 = f_2
        f_2 = f2(x2)
    else:
        b = x2
        x2 = x1
        x1 = a + (b - x2)
        f_2 = f_1
        f_1 = f2(x1)

print('Минимум f2 (Фибоначчи):', (x1 + x2) / 2)
```

### 3. Функция Растригина.

```python
from scipy.optimize import minimize, Bounds
import numpy as np
import matplotlib.pyplot as plt
from math import pi

def func(x):
    return 20 + x[0]**2 + x[1]**2 - 10 * np.cos(2 * pi * x[0]) - 10 * np.cos(2 * pi * x[1])


def func1(x):
    return -func(x)

mins_x = []
mins_y = []
maxs_x = []
maxs_y = []


for i in range(-5, 6):
    for j in range(-5, 6):
        bounds = Bounds([i - 0.5, j - 0.5], [i + 0.5, j + 0.5])  # Устанавливаем локальные границы
        result = minimize(func, (i, j), bounds=bounds)
        mins_x.append(result.x[0])
        mins_y.append(result.x[1])


for i in range(-5, 5):
    for j in range(-5, 5):
        bounds = Bounds([i, j], [i + 1, j + 1])  # Устанавливаем локальные границы
        result = minimize(func1, (i + 0.5, j + 0.5), bounds=bounds)
        maxs_x.append(result.x[0])
        maxs_y.append(result.x[1])


plt.figure(figsize=(8, 8))
plt.scatter(mins_x, mins_y, label='Минимумы', color='blue')
plt.scatter(maxs_x, maxs_y, label='Максимумы', color='red')
plt.title("Минимумы и максимумы функции")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.show()
```

### 4. Градиенты градиентами

### Решение

Первая картинка иллюстрирует метод наискорейшего спуска, так как шаги метода являются ортогональными. Вторая картинка, соответственно, относится к методу сопряженных градиентов. Замкнутые линии на изображениях представляют собой линии уровня, показывающие области, где значение функции остаётся одинаковым.

В случае, если оптимизируемая функция квадратичная и определена на пространстве \(\mathbb{R}^n\), метод сопряженных градиентов находит минимум не более чем за \(n\) шагов.


### Доказательство

Пусть необходимо минимизировать квадратичную функцию:
\[
F(x) = \frac{1}{2}(Ax, x) - (b, x) \rightarrow \inf, \quad x \in \mathbb{R}^n.
\]

Минимизация \(F(x)\) эквивалентна решению системы уравнений:
\[
F'(x) = 0, \quad \text{или} \quad Ax - b = 0.
\]

Решение \(x^*\) можно представить как разложение по некоторому базису:
\[
x^* = x_0 + a_1 p_1 + a_2 p_2 + \dots + a_n p_n,
\]
где:
- \(x_0\) — начальная точка в пространстве \(\mathbb{R}^n\) (задаётся в методе сопряженных градиентов),
- \(p_1, \dots, p_n\) — ортогональные базисные векторы.

Метод сопряженных градиентов строит это разложение, добавляя по одному базисному вектору на каждом шаге. Таким образом, для нахождения решения \(x^*\) требуется не более \(n\) итераций (может потребоваться меньше, если в разложении \(x^*\) задействованы не все базисные векторы).


### 6. Сопрягаемся (полузвёздочка)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def f(x):
    """Целевая функция."""
    return (1/2)*(x[0]**4 - 16*x[0]**2 + 5*x[0] + x[1]**4 - 16*x[1]**2 + 5*x[1])

def grad_f(x):
    """Градиент целевой функции."""
    grad_x0 = (1/2)*(4*x[0]**3 - 32*x[0] + 5)
    grad_x1 = (1/2)*(4*x[1]**3 - 32*x[1] + 5)
    return np.array([grad_x0, grad_x1])

def conjugate_gradients(starting_point, epsilon, param):
    """
    Метод сопряженных градиентов.
    
    :param starting_point: Начальная точка
    :param epsilon: Погрешность
    :param param: Параметр для выбора алгоритма (1 - Флетчера-Ривса, 2 - Полака-Рибьера)
    :return: Точка минимума, количество итераций, шаги
    """
    x = starting_point
    iterations = 0
    steps = [x]
    flag = 0
    r = -grad_f(x)  # Начальный остаток
    S = r           # Направление сопряженного градиента

    while iterations < 1000:
        # Минимизация по направлению сопряженного градиента
        phi = lambda a: f(x + a * S)
        alpha = minimize_scalar(phi).x
        x_new = x + alpha * S

        r_new = -grad_f(x_new)  # Новый остаток
        
        # Вычисление beta для двух методов
        if param == 1:  # Флетчера-Ривса
            beta = np.dot(r_new, r_new) / np.dot(r, r)
        elif param == 2:  # Полака-Рибьера
            beta = max(0, np.dot(r_new, r_new - r) / np.dot(r, r))

        if flag:
            beta = 0  # Остановка на первой итерации

        S = r_new + beta * S  # Обновление направления

        iterations += 1
        steps.append(x_new)

        # Условие остановки: когда разница между новыми и старыми точками меньше epsilon
        if np.linalg.norm(x_new - x) < epsilon:
            if flag:
                x = x_new
                break
            else:
                flag = 1
                x = x_new
                r = r_new
                continue

        flag = 0
        x = x_new
        r = r_new

    return x, iterations, np.array(steps)


# Параметры
epsilon = 1e-6

print('Алгоритм Флетчера-Ривса:')
# Тестирование для разных начальных точек с методом Флетчера-Ривса
starting_points = [np.array([0.1, -1]), np.array([0.1, -1.1]), np.array([0, -1])]
for starting_point in starting_points:
    min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 1)
    print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

print('\nАлгоритм Полака-Рибьера:')
# Тестирование для разных начальных точек с методом Полака-Рибьера
for starting_point in starting_points:
    min_point, iterations, steps = conjugate_gradients(starting_point, epsilon, 2)
    print(f'Start point: {starting_point}, Min point: {min_point}, Iterations: {iterations}')

print('\nВидно, что в данных условиях второй вариант (Полака-Рибьера) работает намного лучше.')
```
