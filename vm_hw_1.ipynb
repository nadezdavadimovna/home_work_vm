```python
import numpy as np
```

```python
M_1 = np.array([[1, 2], [3, 4]])
M_2 = np.array([[5], [6]])
combined_matrix = np.hstack((M_1, M_2)) 

combined_matrix
```

```python
M = np.array([[1, 2], [3, 4]])
a = np.array([0, 1, 0])
b = np.array([1, 0, 0])

result_array = M[a, b]

result_array
```

```python
M = np.array([[1, 2], [3, 4]])
trace = np.trace(M)
nonzero_diag_elements = np.count_nonzero(np.diagonal(M))

trace
```

```python
M_1 = np.array([[1, -1, 3], [-2, -2, 2]])
M_2 = np.array([[1, 2, 3], [4, 5, 6]])

M_2[M_1 < 0] = 0

M_2
```

```python
M_1 = np.array([[1, 2], [3, 4], [5, 6]])
M_2 = np.array([[7, 8], [9, 10], [11, 12]])

matrix_products = np.dot(M_1.T, M_2)

matrix_products
```

**Номер 2**

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 2)

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))
axes[0, 1].plot(x, np.sqrt(x))
axes[1, 0].plot(x, x**2)
axes[1, 1].plot(x, x**3)

plt.show()
```

```python
x = np.linspace(0, 10, 10)
y = np.sqrt(x)
error_y = np.random.rand(10) * 0.2
error_x = np.random.rand(10) * 0.5

plt.errorbar(x, y, yerr=error_y, xerr=error_x, fmt='o', color='red', ecolor='blue', capsize=5, markersize=8)
plt.plot(np.linspace(0, 10, 100), np.sqrt(np.linspace(0, 10, 100)), linestyle='-.', color='green', linewidth=2)

plt.show()
```

```python
x_log = np.logspace(0.5, 1, 100) 
y_log = np.logspace(5, 10, 100) 
plt.loglog(x_log, y_log)
plt.xlabel('X(log scale)')
plt.ylabel('Y(log scale)')

plt.show()
```
